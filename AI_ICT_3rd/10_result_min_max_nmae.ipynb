```python
import pandas as pd
import numpy as np
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_absolute_error
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
result_data = pd.read_csv('../CSV/0712_lgbm_result_lag.csv')
```

```python
import pandas as pd
from sklearn.metrics import mean_absolute_error

# 가정: result_data는 데이터프레임이며 'original_품목번호', '해당일자_전체거래물량', '예측_해당일자_전체거래물량', '해당일자_평균가격', '예측_해당일자_평균가격_shifted' 열이 포함되어 있다.

mae_trade_volume_list = []
min_trade_volume_list = []
max_trade_volume_list = []

mae_avg_price_list = []
min_avg_price_list = []
max_avg_price_list = []

for i in range(1, 30):
    filtered_data = result_data[result_data['original_품목번호'] == i]
    filtered_nrow = filtered_data.shape[0]

    if filtered_nrow > 1:
        df = pd.DataFrame(filtered_data)
        
        mae_trade_volume_data = pd.concat([df['해당일자_전체거래물량'][1:filtered_nrow], df['예측_해당일자_전체거래물량'][1:filtered_nrow]], axis=1).dropna()
        mae_avg_price_data = pd.concat([df['해당일자_평균가격'][1:filtered_nrow], df['예측_해당일자_평균가격_shifted'][1:filtered_nrow]], axis=1).dropna()

        mae_trade_volume = mean_absolute_error(mae_trade_volume_data['해당일자_전체거래물량'], mae_trade_volume_data['예측_해당일자_전체거래물량'])
        min_trade_volume = df['해당일자_전체거래물량'].min()
        max_trade_volume = df['해당일자_전체거래물량'].max()

        mae_avg_price = mean_absolute_error(mae_avg_price_data['해당일자_평균가격'], mae_avg_price_data['예측_해당일자_평균가격_shifted'])
        min_avg_price = df['해당일자_평균가격'].min()
        max_avg_price = df['해당일자_평균가격'].max()
    
        mae_trade_volume_list.append(mae_trade_volume)
        min_trade_volume_list.append(min_trade_volume)
        max_trade_volume_list.append(max_trade_volume)

        mae_avg_price_list.append(mae_avg_price)
        min_avg_price_list.append(min_avg_price)
        max_avg_price_list.append(max_avg_price)

print("MAE Trade Volume List:", mae_trade_volume_list)
print("Min Trade Volume List:", min_trade_volume_list)
print("Max Trade Volume List:", max_trade_volume_list)

print("MAE Avg Price List:", mae_avg_price_list)
print("Min Avg Price List:", min_avg_price_list)
print("Max Avg Price List:", max_avg_price_list)
```

```python
import pandas as pd
from sklearn.metrics import mean_absolute_error

result_dataframe = pd.DataFrame(columns=['original_품목번호', 'min_trade_volume', 'max_trade_volume', 'mae_trade_volume','nmae_trade_volume', 'min_avg_price', 'max_avg_price', 'mae_avg_price', 'nmae_avg_price'])

for i in range(1, 30):
    filtered_data = result_data[result_data['original_품목번호'] == i]
    filtered_nrow = filtered_data.shape[0]

    if filtered_nrow > 1:
        df = pd.DataFrame(filtered_data)
        
        mae_trade_volume_data = pd.concat([df['해당일자_전체거래물량'].iloc[1:], df['예측_해당일자_전체거래물량'].iloc[1:]], axis=1).dropna()
        mae_avg_price_data = pd.concat([df['해당일자_평균가격'].iloc[1:], df['예측_해당일자_평균가격_shifted'].iloc[1:]], axis=1).dropna()

        mae_trade_volume = int(mean_absolute_error(mae_trade_volume_data['해당일자_전체거래물량'], mae_trade_volume_data['예측_해당일자_전체거래물량']))
        min_trade_volume = int(df['해당일자_전체거래물량'].min())
        max_trade_volume = int(df['해당일자_전체거래물량'].max())
        if max_trade_volume == min_trade_volume:
            nmae_trade_volume = 0
        else:
            nmae_trade_volume = mae_trade_volume / (max_trade_volume - min_trade_volume)

        mae_avg_price = int(mean_absolute_error(mae_avg_price_data['해당일자_평균가격'], mae_avg_price_data['예측_해당일자_평균가격_shifted']))
        min_avg_price = int(df['해당일자_평균가격'].min())
        max_avg_price = int(df['해당일자_평균가격'].max())
        if max_avg_price == min_avg_price:
            nmae_avg_price = 0
        else:
            nmae_avg_price = mae_avg_price / (max_avg_price - min_avg_price)

        result_dataframe = result_dataframe.append({'original_품목번호': i,
                                                    'mae_trade_volume': mae_trade_volume,
                                                    'min_trade_volume': min_trade_volume,
                                                    'max_trade_volume': max_trade_volume,
                                                    'mae_avg_price': mae_avg_price,
                                                    'min_avg_price': min_avg_price,
                                                    'max_avg_price': max_avg_price,
                                                    'nmae_trade_volume': nmae_trade_volume,
                                                    'nmae_avg_price': nmae_avg_price,}, ignore_index=True)
```

```python
result_dataframe.set_index('original_품목번호', inplace=True)
result_dataframe.T
```
